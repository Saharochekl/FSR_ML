Основными параметрами для настройки Дерева являются:
1. *criterion* - как считается Прирост Информации
2. *splitter* - как выбирается по какой фиче и порогу фичи производится сплит в узле Дерева

5. *min_samples_leaf* - минимальное требуемое количество сэмплов в узле, чтобы узел мог быть листом
6. *max_leaf_nodes* - максимальное количество листьев в Дереве
