## Задача 3

Как и в случае с линейной моделью, алгоритм построения Дерева позволяет нам посмотреть важность наших фичей. Онако, если в линейных моделях важность фичи считается как вес этой фичи в скалярном произведении, то в Дереве это её информативность

Поскольку Дерево стоится так, чтобы максимизировать прирост информации, то логично считать фичу важной, если она в принципе участвовала в разделении, причем чем большее число раз и больший прирост информации фича дала, тем она и важнее. Таким образом, важность фичи есть суммарный прирост информации по всем узлам, в которых она участвовала. Фича, которая ни разу не использовалась для разделения, имеет нулевую важность

Загрузите датасет *richters_whole_1.csv*

```python
df = pd.read_csv("richters_whole_1.csv")
print(f"Shape: {df.shape}")
print("Columns:", list(df.columns))

display(df.head())

print("\nDtypes:")
print(df.dtypes)
```

Обучите стоковое Дерево и выведите график Важности Фичей

```python

from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
from sklearn.metrics import accuracy_score, f1_score, r2_score, mean_squared_error, classification_report
```

```python

target_col = "damage_grade"
y = df[target_col]
X = df.drop(columns=[target_col])

# Приводим X к числам (если попадутся object/category) и импутация пропусков
X_num = X.copy()
for c in X_num.select_dtypes(include=["object", "category"]).columns:
    X_num[c] = X_num[c].astype("category").cat.codes
    

imp = SimpleImputer(strategy="median")
X_imp = imp.fit_transform(X_num)

print(target_col)
```

```python


# Классификация или регрессия?
is_classif = (str(y.dtype) in ["int8","int16","int32","int64","uint8","uint16","uint32","uint64"]) and (y.nunique() <= 20 or y.dtype.name == "category")

if is_classif:
    X_tr, X_te, y_tr, y_te = train_test_split(X_imp, y, test_size=0.25, random_state=None, stratify=y)
    model = DecisionTreeClassifier(random_state=42)
else:
    X_tr, X_te, y_tr, y_te = train_test_split(X_imp, y, test_size=0.25, random_state=None)
    model = DecisionTreeRegressor(random_state=42)

# Обучаем "стоковое" дерево
model.fit(X_tr, y_tr)

# Быстрая оценка качества
y_pred = model.predict(X_te)
if is_classif:
    acc = accuracy_score(y_te, y_pred)
    f1m = f1_score(y_te, y_pred, average="macro", zero_division=0)
    print(f"DecisionTreeClassifier: acc={acc:.4f}, f1_macro={f1m:.4f}")
else:
    r2 = r2_score(y_te, y_pred)
    rmse = mean_squared_error(y_te, y_pred, squared=False)
    print(f"DecisionTreeRegressor: R2={r2:.4f}, RMSE={rmse:.4f}")

# Важности фичей
importances = model.feature_importances_
feat_names = np.array(X_num.columns)

# Сортировка и top-20
order = np.argsort(importances)[::-1]
topk = min(20, len(feat_names))
imp_vals = importances[order][:topk]
imp_names = feat_names[order][:topk]
```

```python

# График (горизонтальные бары, чтобы имена читались)
plt.figure()
ypos = np.arange(topk)
plt.barh(ypos, imp_vals)
plt.yticks(ypos, imp_names)
plt.gca().invert_yaxis()
plt.title("Decision Tree: feature importances (top-20)")
plt.xlabel("importance")
plt.tight_layout()
plt.show()
```

**Вопрос:** Какие фичи оказались важными? Есть ли неважные фичи? Предположите почему?

Странно, что неизвестная фича оказалась настолько важной. Остальные с уровнем земли в целом достаточно предсказуемые

Произведите подбор гиперпараметров Дерева с помощью [GridSearchCV](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.GridSearchCV.html), оцените качество классификации и выведите график Важности Фичей

```python
from sklearn.model_selection import GridSearchCV, StratifiedKFold

for c in X.select_dtypes(include=["object", "category"]).columns:
    X[c] = X[c].astype("category").cat.codes

# Разбиение и импутация пропусков
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)
imp = SimpleImputer(strategy="median")
X_train = imp.fit_transform(X_train)
X_test  = imp.transform(X_test)
```

```python

# GridSearchCV по дереву
param_grid = {
    "criterion": ["gini", "entropy"],
    "max_depth": [None, 5, 8, 12, 16, 24],
    "min_samples_split": [2, 5, 10, 20],
    "min_samples_leaf": [1, 2, 5, 10],
    "max_features": [None, "sqrt", "log2"]
}
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

gs = GridSearchCV(
    DecisionTreeClassifier(random_state=42),
    param_grid=param_grid,
    scoring="f1_macro",
    cv=cv, n_jobs=-1, return_train_score=True, verbose=0
)
gs.fit(X_train, y_train)

best_tree = gs.best_estimator_
print("Best params:", gs.best_params_, "| best CV f1_macro:", f"{gs.best_score_:.4f}")

# Качество на тесте
y_pred = best_tree.predict(X_test)
acc = accuracy_score(y_test, y_pred)
f1m = f1_score(y_test, y_pred, average="macro", zero_division=0)
print(f"Test: acc={acc:.4f}  f1_macro={f1m:.4f}")
print("\nClassification report:\n", classification_report(y_test, y_pred, zero_division=0))
```

```python

# График важностей после тюнинга
feat_names = np.array(df.drop(columns=[target_col]).columns)
importances = best_tree.feature_importances_
order = np.argsort(importances)[::-1]
k = min(20, len(feat_names))

plt.figure()
plt.barh(range(k), importances[order][:k])
plt.yticks(range(k), feat_names[order][:k])
plt.gca().invert_yaxis()
plt.title("Decision Tree (tuned): feature importances (top-20)")
plt.xlabel("importance")
plt.tight_layout()
plt.show()
```

**Вопрос:** Как изменилась фажность фичей после тюнинга? Предположите почему?

Самая важная фича стала ещё более важной. А в остальном, по факту опеределились самые важные фичи из топ-5. Интересно только увеличение важности has_superstructure_mud_mortar_stone

Постройте модель на важных фичах и произвелите её настройку. Сравните результат с настроенной полной моделью

```python
from sklearn.pipeline import Pipeline
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
pipe = Pipeline([("imp", SimpleImputer(strategy="median")),
                 ("clf", DecisionTreeClassifier(random_state=42))])

param_grid = {
    "clf__criterion": ["gini", "entropy"],
    "clf__max_depth": [None, 5, 8, 12, 16, 24],
    "clf__min_samples_split": [2, 5, 10, 20],
    "clf__min_samples_leaf": [1, 2, 5, 10],
    "clf__max_features": [None, "sqrt", "log2"]
}

# Полная настроенная модель (база для сравнения)
gs_full = GridSearchCV(pipe, param_grid, scoring="f1_macro",
                       cv=cv, n_jobs=-1, return_train_score=True, verbose=0)
gs_full.fit(X_train, y_train)
full_best = gs_full.best_estimator_
y_pred_full = full_best.predict(X_test)

acc_full = accuracy_score(y_test, y_pred_full)
f1m_full = f1_score(y_test, y_pred_full, average="macro", zero_division=0)
print("FULL best params:", gs_full.best_params_, "| CV f1_macro:", f"{gs_full.best_score_:.4f}")
print(f"FULL Test: acc={acc_full:.4f}  f1_macro={f1m_full:.4f}")
print(classification_report(y_test, y_pred_full, zero_division=0))
```

```python
# Берём дерево из пайплайна full_best
tree_full = full_best.named_steps["clf"]

# Имена признаков – те же, что в X (до split'а)
feat_names = np.array(X.columns)

# Важности и сортировка
importances = tree_full.feature_importances_
order = np.argsort(importances)[::-1]

imp_sorted = importances[order]
feat_sorted = feat_names[order]

# Кумулятивная сумма важностей
cum = np.cumsum(imp_sorted)

# Пороги кумулятивной важности
thresholds = [0.95, 0.90, 0.80, 0.70]


best_estimators = {}
sel_features_by_thr = {}
results = []

for thr in thresholds:
    # минимум фич, покрывающих thr долю кумулятивной важности
    k = int(np.searchsorted(cum, thr) + 1)
    k = min(k, len(feat_sorted))

    sel_feats = feat_sorted[:k]
    sel_features_by_thr[thr] = sel_feats

    print(f"\n=== Порог: {int(thr * 100)}% кум. важности ===")
    print(f"Выбрано фич: {k} из {len(feat_names)}")

    gs_sel = GridSearchCV(
        pipe,
        param_grid,
        scoring="f1_macro",
        cv=cv,
        n_jobs=-1,
        return_train_score=True,
        verbose=0,
    )
    gs_sel.fit(X_train[sel_feats], y_train)

    sel_best = gs_sel.best_estimator_
    best_estimators[thr] = sel_best

    y_pred_sel = sel_best.predict(X_test[sel_feats])

    acc_sel = accuracy_score(y_test, y_pred_sel)
    f1m_sel = f1_score(y_test, y_pred_sel, average="macro", zero_division=0)

    print("SEL  best params:", gs_sel.best_params_, "| CV f1_macro:", f"{gs_sel.best_score_:.4f}")
    print(f"SEL  Test: acc={acc_sel:.4f}  f1_macro={f1m_sel:.4f}")
    print(classification_report(y_test, y_pred_sel, zero_division=0))

    results.append({
        "thr": thr,
        "model": f"Selected tuned ({int(thr * 100)}%)",
        "features": len(sel_feats),
        "acc": acc_sel,
        "f1_macro": f1m_sel,
    })
```

```python
summary = pd.DataFrame(
    [{"model": "Full tuned", "features": X.shape[1], "acc": acc_full, "f1_macro": f1m_full}]
    + results
).sort_values("f1_macro", ascending=False)

print("\nСводка:")
print(summary.to_string(index=False))
```

```python
# Оставляем только урезанные модели (results) и ищем среди них лучшую по f1_macro
results_df = pd.DataFrame(results)
best_idx = results_df["f1_macro"].idxmax()
best_thr = results_df.loc[best_idx, "thr"]

print(f"\nЛучший урезанный вариант: {int(best_thr * 100)}% кум. важности")

best_model = best_estimators[best_thr]
best_feats = sel_features_by_thr[best_thr]

# Важности для лучшей урезанной модели
tree_sel = best_model.named_steps["clf"]
imp_sel = tree_sel.feature_importances_
names_sel = np.array(best_feats)  # порядок соответствует колонкам X_train[best_feats]

order2 = np.argsort(imp_sel)[::-1]
k = min(20, len(names_sel))

plt.figure()
plt.barh(range(k), imp_sel[order2][:k])
plt.yticks(range(k), names_sel[order2][:k])
plt.gca().invert_yaxis()
plt.title(f"Decision Tree (best selected, {int(best_thr*100)}%): importances (top-{k})")
plt.xlabel("importance")
plt.tight_layout()
plt.show()
```

## Задача 4

Для задачи регрессии Дерево строится похожим образом, только теперь в качестве предсказания необходимо взять среднее целевое значение всех объектов в листе. Чем точнее это среднее, тем лучше, а значит мы все так же, как и в других регрессионных моделях, хотим уменьшать разброс данных, поэтому, для подсчета Прироста Информации используем соотвествующие функции потерь, например MSE

При этом Регрессионное Дерево остается кусочно-постоянной функцией, а значит оно не сможет приближать гладкие кривые с любой точностью. При этом Дерево так же легко переобучается, если не следить, например, за его глубиной. Тогда Дерево просто превратится в кусочно-постоянную функцию только со значениями целевой переменной

Сгенерируйте данные $X_{1}\sim R\bigg(-\dfrac{3\pi}{2},\dfrac{3\pi}{2}\bigg), Y_{1}=\sin(X_{1})$ и $X_{2}\sim R\bigg(-\dfrac{3\pi}{2},\dfrac{3\pi}{2}\bigg), Y_{2}=-\sin(X_{2})$ каждые размером $N=100$ и объедините данные в один датасет

```python
# Генерация двух синусоид и объединение в общий датасет для регрессии деревом

rng = np.random.default_rng()
N = 100
low, high = -1.5 * np.pi, 1.5 * np.pi

# Набор 1: y =  sin(x)
X1 = rng.uniform(low, high, N)
Y1 = np.sin(X1)

# Набор 2: y = -sin(x)
X2 = rng.uniform(low, high, N)
Y2 = -np.sin(X2)

# В один DataFrame (+ метка источника для удобства визуализации)
df1 = pd.DataFrame({"X": X1, "Y": Y1, "group": 1})
df2 = pd.DataFrame({"X": X2, "Y": Y2, "group": 2})
df = pd.concat([df1, df2], ignore_index=True)

# Перемешаем для честного train/test
df = df.sample(frac=1.0, random_state=None).reset_index(drop=True)

# Массивы для sklearn
X = df[["X"]].to_numpy()   # (2N, 1)
y = df["Y"].to_numpy()     # (2N,)

print("Shape X, y:", X.shape, y.shape)
display(df.head())

# Быстрая визуализация
plt.figure()
for g, sub in df.groupby("group"):
    plt.scatter(sub["X"], sub["Y"], s=12, label=f"group {g}")
plt.title("Объединённый датасет: y = ±sin(x)")
plt.xlabel("X"); plt.ylabel("Y")
plt.legend()
plt.show()
```

Обучите стоковое Дерево для решения задачи регрессии, где $X$ - фича, $Y$ - целевая переменная. Изобразите на графике данные и результат работы Дерева на тренировочной выборке

Теперь посмотрите, как меняется кривая, построенная Деревом, в зависимости от различных параметров модели

```python

from sklearn.metrics import mean_squared_error, r2_score


# Предполагается, что X, y уже созданы (см. предыдущую ячейку).
xs = np.linspace(float(X.min()), float(X.max()), 800).reshape(-1, 1)

def run_family(param_name, values, base=None):
    if base is None:
        base = dict(max_depth=None, min_samples_leaf=1, min_samples_split=2, criterion="squared_error", ccp_alpha=0.0)
    curves = {}
    rows = []
    for v in values:
        params = base.copy()
        params[param_name] = v
        model = DecisionTreeRegressor(random_state=42, **params)
        model.fit(X, y)
        yhat_grid = model.predict(xs)

        # метрики на трейне (чтобы увидеть тенденции)
        yhat_train = model.predict(X)
        rmse = mean_squared_error(y, yhat_train)
        r2   = r2_score(y, yhat_train)

        label = str(v) if v is not None else "None"
        curves[label] = yhat_grid
        rows.append({"param": param_name, "value": label, "RMSE_train": rmse, "R2_train": r2})
    return curves, pd.DataFrame(rows)

def plot_family(title, curves):
    plt.figure()
    plt.scatter(X[:, 0], y, s=12, label="train")
    for lab, yh in curves.items():
        plt.plot(xs[:, 0], yh, linewidth=2, label=lab)
    plt.title(title)
    plt.xlabel("X"); plt.ylabel("Y")
    plt.legend()
    plt.show()

depths = [1, 2, 3, 4, 5, 7, 9, 12, None]
curves_d, tbl_d = run_family("max_depth", depths)
plot_family("DecisionTreeRegressor: влияние max_depth", curves_d)
print("max_depth — метрики на трейне:")
print(tbl_d.sort_values("value").to_string(index=False))

leafs = [1, 5, 20, 50]
curves_l, tbl_l = run_family("min_samples_leaf", leafs)
plot_family("DecisionTreeRegressor: влияние min_samples_leaf", curves_l)
print("\nmin_samples_leaf — метрики на трейне")
print(tbl_l.to_string(index=False))

alphas = [0.0, 1e-4, 5e-4, 1e-3, 5e-3, 1e-2]
curves_a, tbl_a = run_family("ccp_alpha", alphas)
plot_family("DecisionTreeRegressor: влияние ccp_alpha (cost-complexity pruning)", curves_a)
print("\nccp_alpha — метрики на трейне")
print(tbl_a.to_string(index=False))

#    'poisson' не используем, т.к. y может быть < 0
criteria = ["squared_error", "absolute_error", "friedman_mse"]
curves_c, tbl_c = run_family("criterion", criteria)
plot_family("DecisionTreeRegressor: влияние criterion", curves_c)
print("\ncriterion — метрики на трейне")
print(tbl_c.to_string(index=False))
```

Используйте Дерево для построения регрессии, предварительно произведите подбор гиперпараметров, оцените качество модели

```python
from sklearn.model_selection import KFold
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error

# разбиение


y = df["gross_price"]
X = df.drop(columns=["gross_price"])

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=None
)

# Препроцессинг: числа -> median-impute; категории -> most_frequent + OHE
num_cols = X_train.select_dtypes(include=[np.number]).columns.tolist()
cat_cols = [c for c in X_train.columns if c not in num_cols]

transformers = []
if num_cols:
    transformers.append(("num", SimpleImputer(strategy="median"), num_cols))
if cat_cols:
    transformers.append(("cat", Pipeline([
        ("imp", SimpleImputer(strategy="most_frequent")),
        ("ohe", OneHotEncoder(handle_unknown="ignore"))
    ]), cat_cols))

preprocess = ColumnTransformer(transformers=transformers, remainder="drop")

# Модель и сетка гиперпараметров
pipe = Pipeline([
    ("prep", preprocess),
    ("reg", DecisionTreeRegressor(random_state=None))
])

param_grid = {
    "reg__criterion": ["squared_error", "friedman_mse", "absolute_error"],
    "reg__max_depth": [None, 4, 6, 8, 12, 16, 24],
    "reg__min_samples_split": [2, 5, 10, 20, 50],
    "reg__min_samples_leaf": [1, 2, 5, 10, 20],
    "reg__max_features": [None, "sqrt", "log2"],
    "reg__ccp_alpha": [0.0, 1e-4, 1e-3, 1e-2]
}

cv = KFold(n_splits=5, shuffle=True, random_state=None)

gs = GridSearchCV(
    pipe, param_grid=param_grid,
    scoring="neg_root_mean_squared_error",  # минимизируем RMSE
    cv=cv, n_jobs=-1, return_train_score=True, verbose=0
)
gs.fit(X_train, y_train)

best = gs.best_estimator_
print("Best params:", gs.best_params_)
print(f"Best CV RMSE: {-gs.best_score_:.3f}")

y_pred = best.predict(X_test)
rmse = mean_squared_error(y_test, y_pred)
mae  = mean_absolute_error(y_test, y_pred)
r2   = r2_score(y_test, y_pred)
print(f"Test RMSE: {rmse:.3f} | MAE: {mae:.3f} | R2: {r2:.4f}")

# Графики: y_true vs y_pred и остатки
plt.figure()
plt.scatter(y_test, y_pred, s=12)
minv = float(min(y_test.min(), y_pred.min()))
maxv = float(max(y_test.max(), y_pred.max()))
plt.plot([minv, maxv], [minv, maxv])
plt.title("DecisionTreeRegressor: y_true vs y_pred (test)")
plt.xlabel("y_true"); plt.ylabel("y_pred")
plt.tight_layout()
plt.show()

resid = y_test.to_numpy() - y_pred
plt.figure()
plt.scatter(y_pred, resid, s=12)
plt.axhline(0, color="k", linewidth=1)
plt.title("Residuals vs Fitted (test)")
plt.xlabel("y_pred"); plt.ylabel("residuals")
plt.tight_layout()
plt.show()
```

```python
# Линейная регрессия с LassoCV + сравнение с DecisionTreeRegressor

from sklearn.linear_model import LassoCV


# Разделение признаков по типам
num_cols = X_train.select_dtypes(include=[np.number]).columns.tolist()
cat_cols = [c for c in X_train.columns if c not in num_cols]

# Препроцессинг: числа -> median + StandardScaler; категории -> most_frequent + OHE
preprocess = ColumnTransformer([
    ("num", Pipeline([("imp", SimpleImputer(strategy="median")),
                      ("sc", StandardScaler())]), num_cols),
    ("cat", Pipeline([("imp", SimpleImputer(strategy="most_frequent")),
                      ("ohe", OneHotEncoder(handle_unknown="ignore"))]), cat_cols)
], remainder="drop")

# Модель: LassoCV (5-fold), увеличим max_iter для надёжной сходимости
cv = KFold(n_splits=5, shuffle=True, random_state=None)
lasso_pipe = Pipeline([
    ("prep", preprocess),
    ("lin", LassoCV(cv=cv, random_state=None, n_alphas=100, max_iter=5000))
])

# Обучение и качество на тесте
lasso_pipe.fit(X_train, y_train)
y_pred_lasso = lasso_pipe.predict(X_test)

rmse_lasso = mean_squared_error(y_test, y_pred_lasso)
mae_lasso  = mean_absolute_error(y_test, y_pred_lasso)
r2_lasso   = r2_score(y_test, y_pred_lasso)

print(f"LassoCV:  alpha_={lasso_pipe.named_steps['lin'].alpha_:.6g}")
print(f"Test: RMSE={rmse_lasso:.3f}  MAE={mae_lasso:.3f}  R2={r2_lasso:.4f}")

# Сравнение с деревом (если переменная best из предыдущей ячейки существует)
rows = [{"model": "LassoCV (linear)",
         "RMSE": rmse_lasso, "MAE": mae_lasso, "R2": r2_lasso}]

if "best" in globals():
    y_pred_tree = best.predict(X_test)
    rmse_tree = mean_squared_error(y_test, y_pred_tree)
    mae_tree  = mean_absolute_error(y_test, y_pred_tree)
    r2_tree   = r2_score(y_test, y_pred_tree)
    print(f"DecisionTreeRegressor (tuned): Test RMSE={rmse_tree:.3f}  MAE={mae_tree:.3f}  R2={r2_tree:.4f}")

    rows.append({"model": "DecisionTree (tuned)",
                 "RMSE": rmse_tree, "MAE": mae_tree, "R2": r2_tree})

summary = pd.DataFrame(rows).sort_values("RMSE")
print("\nСводка:")
print(summary.to_string(index=False))

# Диагностика: y_true vs y_pred и остатки для Lasso
plt.figure()
plt.scatter(y_test, y_pred_lasso, s=12)
lo = float(min(y_test.min(), y_pred_lasso.min()))
hi = float(max(y_test.max(), y_pred_lasso.max()))
plt.plot([lo, hi], [lo, hi])
plt.title("LassoCV: y_true vs y_pred (test)")
plt.xlabel("y_true"); plt.ylabel("y_pred")
plt.tight_layout(); plt.show()

resid = y_test.to_numpy() - y_pred_lasso
plt.figure()
plt.scatter(y_pred_lasso, resid, s=12)
plt.axhline(0, linewidth=1)
plt.title("LassoCV: Residuals vs Fitted (test)")
plt.xlabel("y_pred"); plt.ylabel("residuals")
plt.tight_layout(); plt.show()

# Топ-коэффициенты Lasso по абсолютной величине (top-20)

feat_names = lasso_pipe.named_steps["prep"].get_feature_names_out()


coef = lasso_pipe.named_steps["lin"].coef_
abs_order = np.argsort(np.abs(coef))[::-1]
k = min(20, len(coef))
top_coef = coef[abs_order][:k]
top_names = np.array(feat_names)[abs_order][:k]

plt.figure()
plt.barh(range(k), top_coef[::-1])
plt.yticks(range(k), top_names[::-1])
plt.title("LassoCV: top-20 коэффициентов (по |coef|)")
plt.xlabel("коэффициент")
plt.tight_layout(); plt.show()
```

Лассо получилась получше по $R^2$ в остальном интересно, что фичи у лассо практически все незначимые

Теперь сравните две модели, построенные только от важных фичей

```python
from sklearn.feature_selection import SelectFromModel


# Препроцессинг (числа: median+scale; категории: most_frequent+OHE)
num_cols = X_train.select_dtypes(include=[np.number]).columns.tolist()
cat_cols = [c for c in X_train.columns if c not in num_cols]

preprocess = ColumnTransformer([
    ("num", Pipeline([("imp", SimpleImputer(strategy="median")),
                      ("sc", StandardScaler())]), num_cols),
    ("cat", Pipeline([("imp", SimpleImputer(strategy="most_frequent")),
                      ("ohe", OneHotEncoder(handle_unknown="ignore"))]), cat_cols)
], remainder="drop")

# Отбор важных фичей на основе LassoCV (embedded selection)
cv = KFold(n_splits=5, shuffle=True, random_state=None)

sel_pipe = Pipeline([
    ("prep", preprocess),
    ("sel", SelectFromModel(LassoCV(cv=cv, n_alphas=100, max_iter=5000, random_state=42), threshold="median"))  # берём все с ненулевыми коэффициентами
])

sel_pipe.fit(X_train, y_train)

# Получим имена выбранных признаков
feat_names = sel_pipe.named_steps["prep"].get_feature_names_out()
mask = sel_pipe.named_steps["sel"].get_support()
sel_feat_names = feat_names[mask]
print(f"Выбрано признаков: {mask.sum()} из {len(feat_names)}")

# Преобразуем train/test в "отобранное пространство"
Xtr_sel = sel_pipe.transform(X_train)
Xte_sel = sel_pipe.transform(X_test)

# Модель 1: LassoCV на выбранных фичах
lasso_sel = LassoCV(cv=cv, n_alphas=100, max_iter=5000, random_state=None)
lasso_sel.fit(Xtr_sel, y_train)
yp_lasso = lasso_sel.predict(Xte_sel)

rmse_lasso = mean_squared_error(y_test, yp_lasso)
mae_lasso  = mean_absolute_error(y_test, yp_lasso)
r2_lasso   = r2_score(y_test, yp_lasso)
print(f"LassoCV (selected): RMSE={rmse_lasso:.3f}  MAE={mae_lasso:.3f}  R2={r2_lasso:.4f}")
```

```python

# Модель 2: DecisionTreeRegressor на выбранных фичах (с кратким тюнингом)
tree = DecisionTreeRegressor(random_state=None)
param_grid = {
    "max_depth": [None, 4, 6, 8, 12, 16],
    "min_samples_split": [2, 5, 10, 20],
    "min_samples_leaf": [1, 2, 5, 10],
    "criterion": ["squared_error", "friedman_mse", "absolute_error"],
    "ccp_alpha": [0.0, 1e-4, 1e-3]
}
gs_tree_sel = GridSearchCV(tree, param_grid, scoring="neg_root_mean_squared_error",cv=cv, n_jobs=-1, verbose=0, return_train_score=True)
gs_tree_sel.fit(Xtr_sel, y_train)
tree_sel = gs_tree_sel.best_estimator_

yp_tree = tree_sel.predict(Xte_sel)
rmse_tree = mean_squared_error(y_test, yp_tree)
mae_tree  = mean_absolute_error(y_test, yp_tree)
r2_tree   = r2_score(y_test, yp_tree)
print(f"DecisionTree (selected): RMSE={rmse_tree:.3f}  MAE={mae_tree:.3f}  R2={r2_tree:.4f}")
print("Best tree params (selected):", gs_tree_sel.best_params_)
```

```python

# Базовые полные модели для ориентира, если есть ранее обученные:
rows = [{"model": "Lasso (selected)", "RMSE": rmse_lasso, "MAE": mae_lasso, "R2": r2_lasso},
        {"model": "Tree  (selected)", "RMSE": rmse_tree,  "MAE": mae_tree,  "R2": r2_tree}]

if "lasso_pipe" in globals():
    ypl = lasso_pipe.predict(X_test)
    rows.append({"model": "Lasso (full)", "RMSE": mean_squared_error(y_test, ypl),
                 "MAE": mean_absolute_error(y_test, ypl), "R2": r2_score(y_test, ypl)})
if "best" in globals():
    ypt = best.predict(X_test)
    rows.append({"model": "Tree  (full)", "RMSE": mean_squared_error(y_test, ypt),
                 "MAE": mean_absolute_error(y_test, ypt), "R2": r2_score(y_test, ypt)})

summary = pd.DataFrame(rows).sort_values("RMSE")
print("\nСводка:")
print(summary.to_string(index=False))
```
