## Задача 3

Как вы заметили в предыдущей задаче, подбор гиперпараметров модели (model tuning) заметно влияет на результат, потому выбор оптимальных параметров так же важен, как и предобраотка данных. Наивный подход, которым вы уже пользовались, заключается в переборе нескольких интересных значений и оценке качества модели на валидационной выборке, такой метод называется Grid Search

Однако выделять каждый раз отдельную валидационную выборку может быть проблемно, особенно когда данных не так много. В таком случае можно воспользоваться кросс-валидацией на тестовой выборке: разбиваем тестовую выборку на n частей, проводим обучение модели на n-1 частях и оцениваем качество на n-ой части, циклически повторям n раз (пока не проведём оценку на всез частях выборки), затем усредняем результат

Для удобной работы в случае подбора нескольких гиперпараметров можно воспользоваться [GridSearchCV](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.GridSearchCV.html), он автоматически осуществляет кросс-валидации и определяет оптимальные параметры из заданного вами диапазона  

Загрузите датасет houses_2.csv, разбейте его на тренировочную и тестовую выборки. Используйте [KNN](https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsClassifier.html) для предсказания параметра Rating, сравните результат с [SVC](https://scikit-learn.org/stable/modules/generated/sklearn.svm.SVC.html) на тестовой выборке

```python
from sklearn.model_selection import train_test_split, StratifiedKFold, GridSearchCV
from sklearn.svm import SVC, SVR
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import  StandardScaler
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    confusion_matrix, classification_report
)
```

```python
df = pd.read_csv("houses_2.csv")
y = df["Rating"]
X = df.drop(columns=["Rating"])


preprocess = Pipeline([("sc", StandardScaler())])

# разбиение и CV
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.25, random_state=42, stratify=y)
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

X_train.shape, X_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn_pipe = Pipeline([
    ("prep", preprocess),
    ("clf", KNeighborsClassifier())
])

knn_grid = {
    "clf__n_neighbors": [3, 5, 11],
    "clf__weights": ["uniform", "distance"],
    "clf__p": [1, 2]  
}

knn_gs = GridSearchCV(
    knn_pipe, knn_grid, scoring="f1_macro", cv=cv, n_jobs=-1, verbose=0
)
knn_gs.fit(X_train, y_train)
knn_best = knn_gs.best_estimator_

print("KNN best params:", knn_gs.best_params_)
y_pred_knn = knn_best.predict(X_test)

acc = accuracy_score(y_test, y_pred_knn)
prec = precision_score(y_test, y_pred_knn, average="macro")
rec = recall_score(y_test, y_pred_knn, average="macro")
f1 = f1_score(y_test, y_pred_knn, average="macro")
print(f"KNN test: acc={acc:.4f}  prec_macro={prec:.4f}  rec_macro={rec:.4f}  f1_macro={f1:.4f}")
print("\nClassification report (KNN):\n", classification_report(y_test, y_pred_knn))

cm = confusion_matrix(y_test, y_pred_knn)
print("Confusion matrix:\n", cm)
```

```python

svc_pipe = Pipeline([
    ("prep", preprocess),
    ("clf", SVC(probability=True, random_state=42))
])

svc_grid = {
    "clf__kernel": ["rbf", "poly", "linear"],
    "clf__C": [0.5, 1, 5, 10],
    "clf__gamma": ["scale", "auto"],   # игнорируется у linear
    "clf__degree": [2, 3]              # игнорируется, если не poly
}

svc_gs = GridSearchCV(
    svc_pipe, svc_grid, scoring="f1_macro", cv=cv, n_jobs=-1, verbose=0
)
svc_gs.fit(X_train, y_train)
svc_best = svc_gs.best_estimator_

print("SVC best params:", svc_gs.best_params_)
y_pred_svc = svc_best.predict(X_test)

acc = accuracy_score(y_test, y_pred_svc)
prec = precision_score(y_test, y_pred_svc, average="macro", zero_division=0)
rec = recall_score(y_test, y_pred_svc, average="macro", zero_division=0)
f1 = f1_score(y_test, y_pred_svc, average="macro", zero_division=0)
print(f"SVC test: acc={acc:.4f}  prec_macro={prec:.4f}  rec_macro={rec:.4f}  f1_macro={f1:.4f}")
print("\nClassification report (SVC):\n", classification_report(y_test, y_pred_svc, zero_division=0))

cm = confusion_matrix(y_test, y_pred_svc)

print("Confusion matrix:\n", cm)
```

```python

rows = []

y_pred_knn = knn_best.predict(X_test)
rows.append({
    "model": "KNN",
    "accuracy": accuracy_score(y_test, y_pred_knn),
    "precision_macro": precision_score(y_test, y_pred_knn, average="macro", zero_division=0),
    "recall_macro": recall_score(y_test, y_pred_knn, average="macro", zero_division=0),
    "f1_macro": f1_score(y_test, y_pred_knn, average="macro", zero_division=0)
})

# SVC
y_pred_svc = svc_best.predict(X_test)
rows.append({
    "model": "SVC",
    "accuracy": accuracy_score(y_test, y_pred_svc),
    "precision_macro": precision_score(y_test, y_pred_svc, average="macro", zero_division=0),
    "recall_macro": recall_score(y_test, y_pred_svc, average="macro", zero_division=0),
    "f1_macro": f1_score(y_test, y_pred_svc, average="macro", zero_division=0)
})

summary = pd.DataFrame(rows).sort_values("f1_macro", ascending=False)
print("Сводное сравнение (тест):\n", summary.to_string(index=False))

best_row = summary.iloc[0]
print(f"\nЛучший по macro-F1: {best_row['model']} "f"(F1={best_row['f1_macro']:.4f}, Acc={best_row['accuracy']:.4f})")
```

Теперь использйте [GridSearchCV](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.GridSearchCV.html) на тренировочной выборке для подбора оптимальных параметров обоих методов. Визуализируйте кривые подбора параметров, и подсветите на них оптимальные параметры ([пример](https://stackoverflow.com/questions/37161563/how-to-graph-grid-scores-from-gridsearchcv))

```python
knn_pipe = Pipeline([("sc", StandardScaler()),
                     ("clf", KNeighborsClassifier())])

k_vals = [1, 2, 3, 5, 7, 11, 15, 21, 31, 41, 51]
param_grid_knn = {
    "clf__n_neighbors": k_vals,
    "clf__weights": ["uniform", "distance"],
    "clf__p": [1, 2],  # L1 / L2
}

gs_knn = GridSearchCV(
    knn_pipe, param_grid_knn,
    scoring="f1_macro", cv=cv, n_jobs=-1, return_train_score=True, verbose=0
)
gs_knn.fit(X_train, y_train)
print("KNN best params:", gs_knn.best_params_, " | best CV f1_macro:", f"{gs_knn.best_score_:.4f}")

# Визуализация: кривые mean_test_score vs k для разных (weights,p)
res_knn = pd.DataFrame(gs_knn.cv_results_)
plt.figure()
for (w, p), grp in res_knn.groupby(["param_clf__weights","param_clf__p"]):
    g = grp.sort_values("param_clf__n_neighbors")
    ks = g["param_clf__n_neighbors"].astype(int).to_numpy()
    sc = g["mean_test_score"].to_numpy()
    plt.plot(ks, sc, marker="o", label=f"w={w}, p={int(p)}")

# Подсветка оптимума
bp = gs_knn.best_params_
best_mask = (res_knn["param_clf__n_neighbors"]==bp["clf__n_neighbors"]) & \
            (res_knn["param_clf__weights"]==bp["clf__weights"]) & \
            (res_knn["param_clf__p"]==bp["clf__p"])
best_row = res_knn.loc[best_mask].iloc[0]

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

svc_pipe = Pipeline([("sc", StandardScaler()),
                     ("clf", SVC(kernel="rbf", probability=True, random_state=42))])

C_range = 10.0 ** np.arange(-4, 4)      # 1e-4 ... 1e3
gamma_range = 10.0 ** np.arange(-4, 4)  # 1e-4 ... 1e3
param_grid_svc = {
    "clf__C": C_range.tolist(),
    "clf__gamma": gamma_range.tolist(),
}

gs_svc = GridSearchCV(
    svc_pipe, param_grid_svc,
    scoring="f1_macro", cv=cv, n_jobs=-1, return_train_score=True, verbose=0
)
gs_svc.fit(X_train, y_train)

plt.scatter(int(best_row["param_clf__n_neighbors"]), best_row["mean_test_score"], s=120, marker="x")

plt.title("KNN GridSearchCV: f1_macro vs k")
plt.xlabel("n_neighbors (k)"); plt.ylabel("mean CV f1_macro")
plt.legend()
plt.grid(True)
plt.show()
```

```python
print("SVC best params:", gs_svc.best_params_, " | best CV f1_macro:", f"{gs_svc.best_score_:.4f}")

# Таблица результатов CV
res_svc = pd.DataFrame(gs_svc.cv_results_).copy()

# Приводим параметры к float (если сетка была числовая)
res_svc["C"] = res_svc["param_clf__C"].astype(float)
res_svc["gamma"] = res_svc["param_clf__gamma"].astype(float)
res_svc["score"] = res_svc["mean_test_score"].astype(float)

# mean_test_score vs C (линии по gamma)
plt.figure()
for gval, grp in res_svc.groupby("gamma"):
    g = grp.sort_values("C")
    plt.semilogx(g["C"].to_numpy(), g["score"].to_numpy(), marker="o", label=f"gamma={gval:g}")
bp = gs_svc.best_params_
plt.scatter(float(bp["clf__C"]), float(bp["clf__gamma"]), s=0)  # фиктивная точка для соответствия типов

# крест в оптимуме
best_mask = (res_svc["C"] == float(bp["clf__C"])) & (res_svc["gamma"] == float(bp["clf__gamma"]))
best_row = res_svc.loc[best_mask].iloc[0]
plt.scatter(best_row["C"], best_row["score"], s=120, marker="x", zorder=5)

plt.title("SVC(RBF) GridSearchCV: f1_macro vs C (линии по gamma)")
plt.xlabel("C (log)"); plt.ylabel("mean CV f1_macro")
plt.grid(True); plt.legend()
plt.show()

# mean_test_score vs gamma (линии по C)
plt.figure()
for cval, grp in res_svc.groupby("C"):
    g = grp.sort_values("gamma")
    plt.semilogx(g["gamma"].to_numpy(), g["score"].to_numpy(), marker="o", label=f"C={cval:g}")

# крест в оптимуме
plt.scatter(best_row["gamma"], best_row["score"], s=120, marker="x", zorder=5)

plt.title("SVC(RBF) GridSearchCV: f1_macro vs gamma (линии по C)")
plt.xlabel("gamma (log)"); plt.ylabel("mean CV f1_macro")
plt.grid(True); plt.legend()
plt.show()
```

Визуализируйте истинную кривую и результат (на тесте), оцените качество модели. Попроуйте улучшить качество, изменяя параметры модели

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

knn_base = KNeighborsRegressor(n_neighbors=11, weights="distance", p=2)
knn_base.fit(X_train, Y_train)
Y_pred_base = knn_base.predict(X_test)

mae_base  = np.mean(np.abs(Y_pred_base - Y_test), axis=0)
rmse_base = np.sqrt(np.mean((Y_pred_base - Y_test)**2, axis=0))
rmse_mean_base = float(np.sqrt(np.mean((Y_pred_base - Y_test)**2)))
r2_base   = r2_score(Y_test, Y_pred_base, multioutput="uniform_average")

print(f"[BASE] n_neighbors=11, weights=distance, p=2")
print(f"  MAE:  x={mae_base[0]:.4f}, y={mae_base[1]:.4f}")
print(f"  RMSE: x={rmse_base[0]:.4f}, y={rmse_base[1]:.4f}, mean={rmse_mean_base:.4f}")
print(f"  R2 (uniform_average) = {r2_base:.4f}\n")

# Перебор гиперпараметров на train
param_grid = {
    "n_neighbors": [3, 5, 7, 11, 15, 21, 30, 50, 100],
    "weights": ["uniform", "distance"],
    "p": [1, 2],  # манхэттен/евклид
}
# усредняем MSE по двум координатам; GridSearchCV сам усреднит по выходам
gs = GridSearchCV(
    KNeighborsRegressor(),
    param_grid=param_grid,
    scoring="neg_mean_squared_error",
    cv=5,
    n_jobs=-1,
    verbose=0,
    return_train_score=True,
)
gs.fit(X_train, Y_train)

best_knn = gs.best_estimator_
print("Best params:", gs.best_params_, "| best CV MSE:", -gs.best_score_)

# Оценка на тесте лучшей модели
Y_pred = best_knn.predict(X_test)
mae  = np.mean(np.abs(Y_pred - Y_test), axis=0)
rmse = np.sqrt(np.mean((Y_pred - Y_test)**2, axis=0))
rmse_mean = float(np.sqrt(np.mean((Y_pred - Y_test)**2)))
r2   = r2_score(Y_test, Y_pred, multioutput="uniform_average")

print(f"\n[BEST] {best_knn}")
print(f"  MAE:  x={mae[0]:.4f}, y={mae[1]:.4f}")
print(f"  RMSE: x={rmse[0]:.4f}, y={rmse[1]:.4f}, mean={rmse_mean:.4f}")
print(f"  R2 (uniform_average) = {r2:.4f}")


t_grid = np.linspace(0, 2*np.pi, 1000).reshape(-1, 1)
curve_true = np.column_stack((np.cos(t_grid)**3, np.sin(t_grid)**3))

# Упорядочим тест по t, чтобы соединить предсказания кривой линией
order = np.argsort(X_test[:, 0])
Xt_sorted = X_test[order]
Yt_sorted = Y_test[order]
Yp_sorted = Y_pred[order]

plt.figure()
# истинная кривая
plt.plot(curve_true[:, 0], curve_true[:, 1], linewidth=2, label="True curve", color="red")
# наблюдения теста
plt.scatter(Y_test[:, 0], Y_test[:, 1], s=10, label="Test points", color="green")
# предсказанная линия на тестовых t
plt.plot(Yp_sorted[:, 0], Yp_sorted[:, 1], linewidth=2, label="KNN prediction (test)", color="blue")
plt.title("KNNRegression: (x,y) ~ t")
plt.xlabel("x"); plt.ylabel("y")
plt.axis("equal")
plt.legend()
plt.show()

# Кривые подбора (по результатам GridSearchCV) — по k для разных настроек
res = pd.DataFrame(gs.cv_results_)
plt.figure()
for (w, p), grp in res.groupby(["param_weights", "param_p"]):
    g = grp.sort_values("param_n_neighbors")
    ks = g["param_n_neighbors"].astype(int).to_numpy()
    score = g["mean_test_score"].to_numpy()  # neg MSE
    plt.plot(ks, -score, marker="o", label=f"weights={w}, p={int(p)}")

bp = gs.best_params_
mask = (res["param_n_neighbors"]==bp["n_neighbors"]) & \
       (res["param_weights"]==bp["weights"]) & \
       (res["param_p"]==bp["p"])
best_row = res.loc[mask].iloc[0]
plt.scatter(int(best_row["param_n_neighbors"]), -best_row["mean_test_score"], s=120, marker="x")

plt.title("KNN GridSearchCV: test MSE vs k")
plt.xlabel("n_neighbors (k)"); plt.ylabel("mean CV MSE")
plt.grid(True); plt.legend()
plt.show()
```

Сравните результат с [SVR](https://scikit-learn.org/stable/modules/generated/sklearn.svm.SVR.html) на тех же данных

```python
from sklearn.multioutput import MultiOutputRegressor


# Pipeline: стандартизация t -> MultiOutput(SVR)
svr_pipe = Pipeline([
    ("sc", StandardScaler()),
    ("svr", MultiOutputRegressor(SVR()))
])

C_range = 10.0 ** np.arange(-2, 3)
gamma_range = ["scale", "auto", 0.1, 1.0] # компактно и быстро
eps_range = [0.01, 0.03, 0.1]

param_grid_svr = {
    "svr__estimator__kernel": ["rbf", "poly", "linear"],
    "svr__estimator__C": C_range.tolist(),
    "svr__estimator__gamma": gamma_range,
    "svr__estimator__epsilon": eps_range,
    "svr__estimator__degree": [2, 3]  # игнорируется, если не poly
}

gs_svr = GridSearchCV(
    svr_pipe, param_grid_svr,
    scoring="neg_mean_squared_error", cv=5,
    n_jobs=-1, return_train_score=True, verbose=0
)
gs_svr.fit(X_train, Y_train)
best_svr = gs_svr.best_estimator_

print("SVR best params:", gs_svr.best_params_, "| best CV MSE:", -gs_svr.best_score_)

# 3) Оценка на тесте
Y_pred_svr = best_svr.predict(X_test)
mae_svr  = np.mean(np.abs(Y_pred_svr - Y_test), axis=0)
rmse_svr = np.sqrt(np.mean((Y_pred_svr - Y_test)**2, axis=0))
rmse_mean_svr = float(np.sqrt(np.mean((Y_pred_svr - Y_test)**2)))
r2_svr   = r2_score(Y_test, Y_pred_svr, multioutput="uniform_average")

print(f"\n[SVR] MAE:  x={mae_svr[0]:.4f}, y={mae_svr[1]:.4f}")
print(f"[SVR] RMSE: x={rmse_svr[0]:.4f}, y={rmse_svr[1]:.4f}, mean={rmse_mean_svr:.4f}")
print(f"[SVR] R2 (uniform_average) = {r2_svr:.4f}")

# Визуализация: истинная кривая + KNN и SVR на тесте 
t_grid = np.linspace(0, 2*np.pi, 1000).reshape(-1, 1)
curve_true = np.column_stack((np.cos(t_grid)**3, np.sin(t_grid)**3))

order = np.argsort(X_test[:, 0])
Yt_sorted = Y_test[order]
Yp_svr_sorted = Y_pred_svr[order]
```
